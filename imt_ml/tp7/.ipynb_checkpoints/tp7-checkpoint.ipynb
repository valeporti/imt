## Exercice 2.[Decision trees]

**1. calculate several decision trees with conventional impurity measures (gini, information,etc.) and default parameters**

```python
from sklearn import tree
import matplotlib.pyplot as plt
import pylab
pylab.rcParams['figure.figsize'] = (16,9)
```

```python
clf_entropy = tree.DecisionTreeClassifier(criterion='entropy')
clf_entropy = clf_entropy.fit(data, target)
t = tree.plot_tree(clf_entropy.fit(data, target),  filled=True)
```

```python
clf_gini = tree.DecisionTreeClassifier(criterion='gini')
clf_gini = clf_gini.fit(data, target)
t = tree.plot_tree(clf_gini.fit(data, target),  filled=True)
```

**2. evaluate and compare their performance (with accuracy for example) and make also com-parison with baseline classifiers (for example with the predict the most common classvalue algorithm and the One Rule algorithm) ; use train and test framework (is there arisk of an overestimated evaluation), and cross validation framework**

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import itertools
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=1)
```

```python
clf_model = tree.DecisionTreeClassifier(criterion='entropy')
clf_model = clf_model.fit(X_train, y_train)
t = tree.plot_tree(clf_model.fit(X_train, y_train),  filled=True)
```

```python
y_pred = clf_model.predict(X_test)
y_test == y_pred
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred, labels=[0, 1, 2])
np.set_printoptions(precision=2)

# Plot normalized confusion matrix
plt.figure(figsize=(4,4))
plot_confusion_matrix(cnf_matrix, classes=target_names, normalize=True, title='Normalized confusion matrix')

plt.show()
print(cnf_matrix)
```

```python
from sklearn.dummy import DummyClassifier
#stratified: most_frequent:prior uniform constant
```

```python
clf_dummy = DummyClassifier(strategy='stratified')
clf_dummy = clf_dummy.fit(X_train, y_train)
```

```python
y_pred_dummy = clf_dummy.predict(X_test)
y_test == y_pred_dummy
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred_dummy))
```

```python
cnf_matrix_dummy = confusion_matrix(y_test, y_pred_dummy, labels=[0, 1, 2])
cnf_matrix_dummy
```

```python
def dummyClass(X_train, y_train, X_test, y_test, strategy):
    clf_dummy = DummyClassifier(strategy=strategy, random_state=1)
    clf_dummy = clf_dummy.fit(X_train, y_train)
    y_pred_dummy = clf_dummy.predict(X_test)
    print(f'------- Strategy: { strategy }')
    print("Accuracy:", metrics.accuracy_score(y_test, y_pred_dummy))
    cnf_matrix_dummy = confusion_matrix(y_test, y_pred_dummy, labels=[0, 1, 2])
    plt.figure(figsize=(4,4))
    plot_confusion_matrix(cnf_matrix_dummy, classes=target_names, normalize=True, title='Normalized confusion matrix')
    plt.show()
    
strategies = ['stratified', 'most_frequent', 'prior', 'uniform']
for s in strategies:
    dummyClass(X_train, y_train, X_test, y_test, s)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
clf_entropy = tree.DecisionTreeClassifier(criterion='entropy')
scores = cross_val_score(clf_entropy, data, target, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
clf_entropy = tree.DecisionTreeClassifier(criterion='gini')
scores = cross_val_score(clf_entropy, data, target, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
clf_entropy = tree.DecisionTreeClassifier(criterion='entropy', min_samples_split=4)
scores = cross_val_score(clf_entropy, data, target, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

> Compared to Decision trees, the baseline classifiers perform notably worse. Fact that let us know that decision trees are at least better than a random guess.

**3. analyse some training parameters like the minimum number of observations that must exist in a node in order for a split to be attempted, the minimum number of observationsin any terminal leaf node and the cp complexity parameter in CART.**

```python
pylab.rcParams['figure.figsize'] = (16,9)
```

```python
clf_entropy = tree.DecisionTreeClassifier(criterion='entropy', min_samples_split=4)
clf_entropy = clf_entropy.fit(data, target)
t = tree.plot_tree(clf_entropy.fit(data, target),  filled=True)

scores = cross_val_score(clf_entropy, data, target, cv=5)
print("Accuracy kfold: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
clf_entropy = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=4)
clf_entropy = clf_entropy.fit(data, target)
t = tree.plot_tree(clf_entropy.fit(data, target),  filled=True)

scores = cross_val_score(clf_entropy, data, target, cv=5)
print("Accuracy kfold: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
clf_entropy = tree.DecisionTreeClassifier(criterion='entropy', max_depth=2)
clf_entropy = clf_entropy.fit(data, target)
t = tree.plot_tree(clf_entropy.fit(data, target),  filled=True)

scores = cross_val_score(clf_entropy, data, target, cv=5)
print("Accuracy kfold: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

> We can notice that playing with parameters could make or not a difference in the accuracy and that even if we change different parameters, the result could be the same. This makes us think that problems could be apporached differently depending on the desired output. 
Also, If we prune the tree by "max_depth", we can adjust the underfitting and overfitting problems.
