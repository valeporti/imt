```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
from statsmodels.graphics.mosaicplot import mosaic
```

```python
tips_data = pd.read_csv('./data/tips.csv', header=None, index_col=0, names=["totbill","tip","sex","smoker","day","time", "size"])
```

(sex of person paying for the meal, 0=male, 1=female)=sex", "(smoker in party?
0=No, 1=Yes)=smoker", "(3=Thur, 4=Frid, 5=Sat, 6=Sun)=day", "(0=Day, 1=Night)=time", 
(size of the party)=size

> QUESTION: Pourquoi mettre les valeurs à ces valeurs nominales?

```python
tips_data['sex'] = tips_data['sex'].replace({ 0: 'male', 1: 'female' })
tips_data['smoker'] = tips_data['smoker'].replace({ 0: 'No', 1: 'Yes' })
tips_data['day'] = tips_data['day'].replace({ 3: 'Thur', 4: 'Frid', 5: 'Sat', 6: 'Sun' })
tips_data['time'] = tips_data['time'].replace({ 0: 'Day', 1: 'Night' })
tips_data
```

```python
tips_data['tip_rate'] = 100 * tips_data['tip'] / tips_data['totbill']
tips_data['tot_per_person'] = tips_data['totbill'] / tips_data['size']
tips_data.round({"tip_rate":2, "tot_per_person":2}) 
tips_data
```
