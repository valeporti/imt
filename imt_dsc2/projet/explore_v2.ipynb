----

### Do the substraction of abilities (aggregate mode)

> <span style="color:tomato"><b>Here, by doing the difference of the summ of the used abilities from team 2 (looser) and Team A (winner), we expect to understand the most used skills for the winning side, as an aggregate (since it's the sum)</b></span>


```python
df10_sample_diff = df10_sample.iloc[:,:10].subtract(df10_sample.iloc[:,10:20].values)
df10_sample_diff_agg = df10_sample_diff.agg("sum", axis="rows")
df10_diff_all = pd.DataFrame(np.matrix([df10_sample_diff_agg]).T, columns=['Usage'])
plot = df10_diff_all.plot(kind='bar')
#plot = plot.get_figure()
#df10_01_twoteams
#df10_diff_all
```

> <span style="color:tomato"><b>Negative values mean that the losing team used that ability the most, positive are the abilities the winners used the most. We can say then, that, as an overall, habilities 2, 3, and 9 are the most decisive in order to win.</b></span>

-------

----

### Do the substraction of abilities (existance mode: zeros or ones)

> <span style="color:tomato"><b>By just takng into account the existence of the used or not used abilities, we just care about whether it has been chosen or not, not by how much, used more as an indicator</b></span>


```python
df10_sample_01_diff = df10_sample_01.iloc[:,:10].subtract(df10_sample_01.iloc[:,10:20].values)
df10_sample_01_diff_agg = df10_sample_01_diff.agg("sum", axis="rows")
df10_01_diff_all = pd.DataFrame(np.matrix([df10_sample_01_diff_agg]).T, columns=['Usage'])
plot = df10_01_diff_all.plot(kind='bar')
#plot = plot.get_figure()
#df10_01_twoteams
#df10_01_diff_all
```

> <span style="color:tomato"><b>From this result, and comparing with the one before, we can see that for the ability 4, it's frequently used for loser teams than for winner teams, and it could be highly correlated with losing, for 5 and 6, it's more often used by losers and highly repeated by them ... </b></span>
