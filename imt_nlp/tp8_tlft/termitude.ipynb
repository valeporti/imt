```python
import re
import os
```

```python
f_irma_path = './data/irma.txt'
f_irma_conll_path = './data/irma.surf.conll.txt'
```

```python
f_irma = open(f_irma_path); f_irma_conll = open(f_irma_conll_path)
limit = 2; article_n = 0; phrase_n = 0
phrase_type_seq = []
text_dict = {}

re_article_delimiter = r'[=]{2,}'
re_structure = r'^(\d+)\t(.+)\t(.+)\t(.+)\t(.+)\t(.+)\t(.+)\t(.+)\t(.+)\t(.+)'

for line in f_irma_conll:
    if limit < article_n: break
    if re.search(re_article_delimiter, line): # find new article
        article_n += 1
        continue
    m = re.match(re_structure, line)
    if not m: continue
    #print(m.group(7), m.group(8))
    word_n = m.group(1)
    if word_n == '1': 
        phrase_n += 1; phrase_type_seq = ''; phrase = ''
    word_orig = m.group(2)
    word_root = m.group(3)
    word_type = m.group(4)
    word_spec = m.group(5)
    word_desc = m.group(6)
    word_dep_n = m.group(7)
    word_dep_type = m.group(8)
    if not f'article{ article_n }' in text_dict: text_dict[ f'article{ article_n }' ] = {}
    if not f'phrase{ phrase_n }' in text_dict[ f'article{ article_n }' ]: text_dict[ f'article{ article_n }' ][ f'phrase{ phrase_n }' ] = {}
    phrase = f'{phrase} {word_orig}' if len(phrase_type_seq) > 0 else f'{word_orig}'
    phrase_type_seq = f'{phrase_type_seq} {word_spec}{word_n}' if len(phrase_type_seq) > 0 else f'{word_spec}{word_n}'
    text_dict[ f'article{ article_n }' ][ f'phrase{ phrase_n }' ]['phrase_type_sequence'] = phrase_type_seq
    text_dict[ f'article{ article_n }' ][ f'phrase{ phrase_n }' ]['phrase'] = phrase

text_dict
```
